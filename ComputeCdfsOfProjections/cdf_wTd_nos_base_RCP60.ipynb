```python
# compute cumulative probability distributions
cdfnum = 1000
cdfstep = int(len(SL_wTd_nos_base_RCP60[:,0])/cdfnum)
print(cdfstep)

SL_wTd_nos_base_R0_RCP60_cdf = [0] * (cdfnum+1)
for t in range(len(SL_wTd_nos_base_RCP60[1,:])):
    print("R0: ",t)
    sortind = np.argsort(SL_wTd_nos_base_RCP60[:,t])

    slcdf = []
    slcdf.append(0)
    for i in range(1,cdfnum+1):
        cdfval = float(i/cdfnum)
        slval = SL_wTd_nos_base_RCP60[sortind[i*cdfstep],t]
        slcdf.append(slval)
    SL_wTd_nos_base_R0_RCP60_cdf=np.vstack([SL_wTd_nos_base_R0_RCP60_cdf, slcdf])


SL_wTd_nos_base_R1_RCP60_cdf = [0] * (cdfnum+1)
for t in range(len(SL_wTd_nos_base_R1_RCP60[1,:])):
    print("R1: ",t)
    sortind = np.argsort(SL_wTd_nos_base_R1_RCP60[:,t])

    slcdf = []
    slcdf.append(0)
    for i in range(1,cdfnum+1):
        cdfval = float(i/cdfnum)
        slval = SL_wTd_nos_base_R1_RCP60[sortind[i*cdfstep],t]
        slcdf.append(slval)
    SL_wTd_nos_base_R1_RCP60_cdf=np.vstack([SL_wTd_nos_base_R1_RCP60_cdf, slcdf])

SL_wTd_nos_base_R2_RCP60_cdf = [0] * (cdfnum+1)
for t in range(len(SL_wTd_nos_base_R2_RCP60[1,:])):
    print("R2: ",t)
    sortind = np.argsort(SL_wTd_nos_base_R2_RCP60[:,t])

    slcdf = []
    slcdf.append(0)
    for i in range(1,cdfnum+1):
        cdfval = float(i/cdfnum)
        slval = SL_wTd_nos_base_R2_RCP60[sortind[i*cdfstep],t]
        slcdf.append(slval)
    SL_wTd_nos_base_R2_RCP60_cdf=np.vstack([SL_wTd_nos_base_R2_RCP60_cdf, slcdf])

SL_wTd_nos_base_R3_RCP60_cdf = [0] * (cdfnum+1)
for t in range(len(SL_wTd_nos_base_R3_RCP60[1,:])):
    print("R3: ",t)
    sortind = np.argsort(SL_wTd_nos_base_R3_RCP60[:,t])

    slcdf = []
    slcdf.append(0)
    for i in range(1,cdfnum+1):
        cdfval = float(i/cdfnum)
        slval = SL_wTd_nos_base_R3_RCP60[sortind[i*cdfstep],t]
        slcdf.append(slval)
    SL_wTd_nos_base_R3_RCP60_cdf=np.vstack([SL_wTd_nos_base_R3_RCP60_cdf, slcdf])

SL_wTd_nos_base_R4_RCP60_cdf = [0] * (cdfnum+1)
for t in range(len(SL_wTd_nos_base_R4_RCP60[1,:])):
    print("R4: ",t)
    sortind = np.argsort(SL_wTd_nos_base_R4_RCP60[:,t])

    slcdf = []
    slcdf.append(0)
    for i in range(1,cdfnum+1):
        cdfval = float(i/cdfnum)
        slval = SL_wTd_nos_base_R4_RCP60[sortind[i*cdfstep],t]
        slcdf.append(slval)
    SL_wTd_nos_base_R4_RCP60_cdf=np.vstack([SL_wTd_nos_base_R4_RCP60_cdf, slcdf])

SL_wTd_nos_base_R5_RCP60_cdf = [0] * (cdfnum+1)
for t in range(len(SL_wTd_nos_base_R5_RCP60[1,:])):
    print("R5: ",t)
    sortind = np.argsort(SL_wTd_nos_base_R5_RCP60[:,t])

    slcdf = []
    slcdf.append(0)
    for i in range(1,cdfnum+1):
        cdfval = float(i/cdfnum)
        slval = SL_wTd_nos_base_R5_RCP60[sortind[i*cdfstep],t]
        slcdf.append(slval)
    SL_wTd_nos_base_R5_RCP60_cdf=np.vstack([SL_wTd_nos_base_R5_RCP60_cdf, slcdf])


Percentile = np.arange(0,float((cdfnum+1)/cdfnum),float(1/cdfnum))

# write cdfs
ncfile = nc.Dataset('Cdfs/SL_wTd_nos_base_RCP60_cdf.nc','w', format='NETCDF4')
ncfile.createDimension('Time', None)
ncfile.createDimension('Percentile', None)

SL_wTd_weighted_base_R0 = ncfile.createVariable('Antarctica', 'f4', ('Time','Percentile'))
SL_wTd_weighted_base_R1 = ncfile.createVariable('EAIS', 'f4', ('Time','Percentile'))
SL_wTd_weighted_base_R2 = ncfile.createVariable('Ross', 'f4', ('Time','Percentile'))
SL_wTd_weighted_base_R3 = ncfile.createVariable('Amundsen', 'f4', ('Time','Percentile'))
SL_wTd_weighted_base_R4 = ncfile.createVariable('Weddell', 'f4', ('Time','Percentile'))
SL_wTd_weighted_base_R5 = ncfile.createVariable('Peninsula', 'f4', ('Time','Percentile'))
p = ncfile.createVariable('Percentile', 'f4', 'Percentile')
t = ncfile.createVariable('Time', 'f4', 'Time')

t[:] = Time
t.units = 'years'
SL_wTd_weighted_base_R0[:,:] = SL_wTd_nos_base_R0_RCP60_cdf
SL_wTd_weighted_base_R1[:,:] = SL_wTd_nos_base_R1_RCP60_cdf
SL_wTd_weighted_base_R2[:,:] = SL_wTd_nos_base_R2_RCP60_cdf
SL_wTd_weighted_base_R3[:,:] = SL_wTd_nos_base_R3_RCP60_cdf
SL_wTd_weighted_base_R4[:,:] = SL_wTd_nos_base_R4_RCP60_cdf
SL_wTd_weighted_base_R5[:,:] = SL_wTd_nos_base_R5_RCP60_cdf
p[:] = Percentile

SL_wTd_weighted_base_R0.units = 'meter'
SL_wTd_weighted_base_R1.units = 'meter'
SL_wTd_weighted_base_R2.units = 'meter'
SL_wTd_weighted_base_R3.units = 'meter'
SL_wTd_weighted_base_R4.units = 'meter'
SL_wTd_weighted_base_R5.units = 'meter'

p.units = 'percent'

ncfile.close()
```
