```python
# FETI_VUB = fETISh
# Experiment BM01
fname = "larmip_data/fETISh/timeseriesA1.txt" # File to read
FETI_VUB_BM01_R1 = []
FETI_VUB_BM01_R2 = []
FETI_VUB_BM01_R3 = []
FETI_VUB_BM01_R4 = []
FETI_VUB_BM01_R5 = []

FETI_VUB_BM01_R1.append(0.0)
FETI_VUB_BM01_R2.append(0.0)
FETI_VUB_BM01_R3.append(0.0)
FETI_VUB_BM01_R4.append(0.0)
FETI_VUB_BM01_R5.append(0.0)

with open(fname) as f:
#    GRIS_LSC_BM01_R1 = [float(row.split(" ")[1]) for row in f]
    for line in f:
        line=line[9:]
        dump=line.split("   ")
        FETI_VUB_BM01_R1.append(float(dump[0]))
        FETI_VUB_BM01_R2.append(float(dump[1]))
        FETI_VUB_BM01_R3.append(float(dump[2]))
        FETI_VUB_BM01_R4.append(float(dump[3]))
        FETI_VUB_BM01_R5.append(float(dump[4]))

FETI_VUB_BM01_R1=np.array(FETI_VUB_BM01_R1)
FETI_VUB_BM01_R2=np.array(FETI_VUB_BM01_R2)
FETI_VUB_BM01_R3=np.array(FETI_VUB_BM01_R3)
FETI_VUB_BM01_R4=np.array(FETI_VUB_BM01_R4)
FETI_VUB_BM01_R5=np.array(FETI_VUB_BM01_R5)

FETI_VUB_BM01_SU=FETI_VUB_BM01_R1+FETI_VUB_BM01_R2+FETI_VUB_BM01_R3+FETI_VUB_BM01_R4+FETI_VUB_BM01_R5


# create and write response function
RF_FETI_VUB_BM01_R1=np.diff(FETI_VUB_BM01_R1)/1
RF_FETI_VUB_BM01_R2=np.diff(FETI_VUB_BM01_R2)/1
RF_FETI_VUB_BM01_R3=np.diff(FETI_VUB_BM01_R3)/1
RF_FETI_VUB_BM01_R4=np.diff(FETI_VUB_BM01_R4)/1
RF_FETI_VUB_BM01_R5=np.diff(FETI_VUB_BM01_R5)/1
RF_FETI_VUB_BM01_SU=np.diff(FETI_VUB_BM01_SU)/1


# plot
plt.plot(FETI_VUB_BM01_R1,label="FETI_VUB_BM01_R1")
plt.plot(FETI_VUB_BM01_R2,label="FETI_VUB_BM01_R2")
plt.plot(FETI_VUB_BM01_R3,label="FETI_VUB_BM01_R3")
plt.plot(FETI_VUB_BM01_R4,label="FETI_VUB_BM01_R4")
plt.plot(FETI_VUB_BM01_R5,label="FETI_VUB_BM01_R5")
plt.plot(FETI_VUB_BM01_SU,label="FETI_VUB_BM01_SU")
plt.legend()

print(len(RF_FETI_VUB_BM01_R1))
```
