```python
# IMAU_VUB = IMAUICE
# Experiment BM32

fname = "larmip_data/IMAUICE/LARMIP_IMAUICE32_m32.dat" # File to read   
with open(fname) as f:
    IMAU_VUB_BM32_time=[]
    IMAU_VUB_BM32_R0=[]
    IMAU_VUB_BM32_R1=[]
    IMAU_VUB_BM32_R2=[]
    IMAU_VUB_BM32_R3=[]
    IMAU_VUB_BM32_R4=[]
    IMAU_VUB_BM32_R5=[]
    
    IMAU_VUB_BM32_time.append(0.0)
    IMAU_VUB_BM32_R0.append(0.0)
    IMAU_VUB_BM32_R1.append(0.0)
    IMAU_VUB_BM32_R2.append(0.0)
    IMAU_VUB_BM32_R3.append(0.0)
    IMAU_VUB_BM32_R4.append(0.0)
    IMAU_VUB_BM32_R5.append(0.0)

    for line in f.readlines():
        #print(line)
        dump=line.split("\t")
        IMAU_VUB_BM32_R0.append(float(dump[7]))
        IMAU_VUB_BM32_R1.append(float(dump[2]))
        IMAU_VUB_BM32_R2.append(float(dump[3]))
        IMAU_VUB_BM32_R3.append(float(dump[4]))
        IMAU_VUB_BM32_R4.append(float(dump[5]))
        IMAU_VUB_BM32_R5.append(float(dump[6]))
        IMAU_VUB_BM32_time.append(float(dump[0]))
IMAU_VUB_BM32_R0=np.array(IMAU_VUB_BM32_R0[1:])
IMAU_VUB_BM32_R1=np.array(IMAU_VUB_BM32_R1[1:])
IMAU_VUB_BM32_R2=np.array(IMAU_VUB_BM32_R2[1:])
IMAU_VUB_BM32_R3=np.array(IMAU_VUB_BM32_R3[1:])
IMAU_VUB_BM32_R4=np.array(IMAU_VUB_BM32_R4[1:])
IMAU_VUB_BM32_R5=np.array(IMAU_VUB_BM32_R5[1:])
IMAU_VUB_BM32_time=np.array(IMAU_VUB_BM32_time[1:])

IMAU_VUB_BM32_SU = IMAU_VUB_BM32_R1+IMAU_VUB_BM32_R2+IMAU_VUB_BM32_R3+IMAU_VUB_BM32_R4+IMAU_VUB_BM32_R5


# create and write response function
RF_IMAU_VUB_BM32_R0=np.diff(IMAU_VUB_BM32_R0)/np.diff(IMAU_VUB_BM32_time)/32
RF_IMAU_VUB_BM32_R1=np.diff(IMAU_VUB_BM32_R1)/np.diff(IMAU_VUB_BM32_time)/32
RF_IMAU_VUB_BM32_R2=np.diff(IMAU_VUB_BM32_R2)/np.diff(IMAU_VUB_BM32_time)/32
RF_IMAU_VUB_BM32_R3=np.diff(IMAU_VUB_BM32_R3)/np.diff(IMAU_VUB_BM32_time)/32
RF_IMAU_VUB_BM32_R4=np.diff(IMAU_VUB_BM32_R4)/np.diff(IMAU_VUB_BM32_time)/32
RF_IMAU_VUB_BM32_R5=np.diff(IMAU_VUB_BM32_R5)/np.diff(IMAU_VUB_BM32_time)/32
RF_IMAU_VUB_BM32_SU=np.diff(IMAU_VUB_BM32_SU)/np.diff(IMAU_VUB_BM32_time)/32


# plot
plt.plot(IMAU_VUB_BM32_R0,label="IMAU_VUB_BM32_R0")
plt.plot(IMAU_VUB_BM32_R1,label="IMAU_VUB_BM32_R1")
plt.plot(IMAU_VUB_BM32_R2,label="IMAU_VUB_BM32_R2")
plt.plot(IMAU_VUB_BM32_R3,label="IMAU_VUB_BM32_R3")
plt.plot(IMAU_VUB_BM32_R4,label="IMAU_VUB_BM32_R4")
plt.plot(IMAU_VUB_BM32_R5,label="IMAU_VUB_BM32_R5")
plt.plot(IMAU_VUB_BM32_SU,label="IMAU_VUB_BM32_SU")
plt.legend()

print(len(RF_IMAU_VUB_BM32_R0))
```
