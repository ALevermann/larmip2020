```python
# MALI_LAN = MALI_LANL
# BM08

fname = "larmip_data/MALI_LANL/SLR_EAIS.txt" # File to read   
with open(fname) as f:
    MALI_LAN_BM08_R1 = np.array([float(row[:]) for row in f])
print(MALI_LAN_BM08_R1[-1],MALI_LAN_Control[-1])
MALI_LAN_BM08_R1=MALI_LAN_BM08_R1-MALI_LAN_Control
print(MALI_LAN_BM08_R1[-1])

fname = "larmip_data/MALI_LANL/SLR_ROSS.txt" # File to read   
with open(fname) as f:
    MALI_LAN_BM08_R2 = np.array([float(row[:]) for row in f])
print(MALI_LAN_BM08_R2[-1],MALI_LAN_Control[-1])
MALI_LAN_BM08_R2=MALI_LAN_BM08_R2-MALI_LAN_Control
print(MALI_LAN_BM08_R2[-1])

fname = "larmip_data/MALI_LANL/SLR_AMUNDSEN.txt" # File to read   
with open(fname) as f:
    MALI_LAN_BM08_R3 = np.array([float(row[:]) for row in f])
print(MALI_LAN_BM08_R3[-1],MALI_LAN_Control[-1])
MALI_LAN_BM08_R3=MALI_LAN_BM08_R3-MALI_LAN_Control
print(MALI_LAN_BM08_R3[-1])

fname = "larmip_data/MALI_LANL/SLR_WEDDELL.txt" # File to read   
with open(fname) as f:
    MALI_LAN_BM08_R4 = np.array([float(row[:]) for row in f])
print(MALI_LAN_BM08_R4[-1],MALI_LAN_Control[-1])
MALI_LAN_BM08_R4=MALI_LAN_BM08_R4-MALI_LAN_Control
print(MALI_LAN_BM08_R4[-1])

fname = "larmip_data/MALI_LANL/SLR_PENINSULA.txt" # File to read   
with open(fname) as f:
    MALI_LAN_BM08_R5 = np.array([float(row[:]) for row in f])
print(MALI_LAN_BM08_R5[-1],MALI_LAN_Control[-1])
MALI_LAN_BM08_R5=MALI_LAN_BM08_R5-MALI_LAN_Control
print(MALI_LAN_BM08_R5[-1])

MALI_LAN_BM08_SU = MALI_LAN_BM08_R1+MALI_LAN_BM08_R2+MALI_LAN_BM08_R3+MALI_LAN_BM08_R4+MALI_LAN_BM08_R5
print(MALI_LAN_BM08_SU[-1])


# create and write response function
RF_MALI_LAN_BM08_R1=np.diff(MALI_LAN_BM08_R1)/8
RF_MALI_LAN_BM08_R2=np.diff(MALI_LAN_BM08_R2)/8
RF_MALI_LAN_BM08_R3=np.diff(MALI_LAN_BM08_R3)/8
RF_MALI_LAN_BM08_R4=np.diff(MALI_LAN_BM08_R4)/8
RF_MALI_LAN_BM08_R5=np.diff(MALI_LAN_BM08_R5)/8
RF_MALI_LAN_BM08_SU=np.diff(MALI_LAN_BM08_SU)/8


# plotting
plt.plot(MALI_LAN_BM08_R1,label="MALI_LAN_BM08_R1")
plt.plot(MALI_LAN_BM08_R2,label="MALI_LAN_BM08_R2")
plt.plot(MALI_LAN_BM08_R3,label="MALI_LAN_BM08_R3")
plt.plot(MALI_LAN_BM08_R4,label="MALI_LAN_BM08_R4")
plt.plot(MALI_LAN_BM08_R5,label="MALI_LAN_BM08_R5")
plt.plot(MALI_LAN_BM08_SU,label="MALI_LAN_BM08_SU")
plt.legend()
```

```python
# Response functions
np.savetxt("RF_MALI_LAN_BM08_R1.dat", RF_MALI_LAN_BM08_R1, delimiter=",")
np.savetxt("RF_MALI_LAN_BM08_R2.dat", RF_MALI_LAN_BM08_R2, delimiter=",")
np.savetxt("RF_MALI_LAN_BM08_R3.dat", RF_MALI_LAN_BM08_R3, delimiter=",")
np.savetxt("RF_MALI_LAN_BM08_R4.dat", RF_MALI_LAN_BM08_R4, delimiter=",")
np.savetxt("RF_MALI_LAN_BM08_R5.dat", RF_MALI_LAN_BM08_R5, delimiter=",")
np.savetxt("RF_MALI_LAN_BM08_SU.dat", RF_MALI_LAN_BM08_SU, delimiter=",")
```
