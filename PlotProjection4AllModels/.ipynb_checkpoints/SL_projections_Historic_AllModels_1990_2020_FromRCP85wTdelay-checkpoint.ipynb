```python
import numpy as np
import netCDF4 as nc
import matplotlib.pylab as plt
import imp
import csv
import pandas as pd
from io import StringIO
```

```python
fname = "../ObservationalData/imbie_dataset-2018_06_08.dat" # File to read
df = pd.read_csv(fname,sep='\t',index_col=-1,header=None)
ObsData = np.array(df.values)
ObsTime = ObsData[:,0]
ObsSL = ObsData[:,3]
ObsUncertainty = ObsData[:,4]
#print(ObsData[:,0])
plt.plot(ObsTime,ObsSL,'k')
plt.plot(ObsTime,ObsSL-ObsUncertainty,':k')
plt.plot(ObsTime,ObsSL+ObsUncertainty,':k')

#ObsDSL = ObsData[:,4]
```

```python
Refyear = 1992
Refyearind = Refyear-1900
Finyear = 2017
Finyearind = Finyear-1900
EnsembleNumber = 20000
YearNumber = 200
```

```python
# Read data
# CISM_NCA

fname="../\ComputeProjection4OneIceModel\EnsembleSingleModelProjections/SL_wTdelay_CISM_NCA_RCP85.nc"
ncf = nc.Dataset(fname, "r")

Time = ncf.variables["Time"][:]
SL_wTdelay_CISM_NCA_R0_RCP85 = ncf.variables["Antarctica"][:]
SL_wTdelay_CISM_NCA_R1_RCP85 = ncf.variables["EAIS"][:]
SL_wTdelay_CISM_NCA_R2_RCP85 = ncf.variables["Ross"][:]
SL_wTdelay_CISM_NCA_R3_RCP85 = ncf.variables["Amundsen"][:]
SL_wTdelay_CISM_NCA_R4_RCP85 = ncf.variables["Weddell"][:]
SL_wTdelay_CISM_NCA_R5_RCP85 = ncf.variables["Peninsula"][:]

# Normalizing
for i in range(0,EnsembleNumber):
    SL_wTdelay_CISM_NCA_R0_RCP85_offset = SL_wTdelay_CISM_NCA_R0_RCP85[i,Refyearind]
    SL_wTdelay_CISM_NCA_R1_RCP85_offset = SL_wTdelay_CISM_NCA_R1_RCP85[i,Refyearind]
    SL_wTdelay_CISM_NCA_R2_RCP85_offset = SL_wTdelay_CISM_NCA_R2_RCP85[i,Refyearind]
    SL_wTdelay_CISM_NCA_R3_RCP85_offset = SL_wTdelay_CISM_NCA_R3_RCP85[i,Refyearind]
    SL_wTdelay_CISM_NCA_R4_RCP85_offset = SL_wTdelay_CISM_NCA_R4_RCP85[i,Refyearind]
    SL_wTdelay_CISM_NCA_R5_RCP85_offset = SL_wTdelay_CISM_NCA_R5_RCP85[i,Refyearind]
    for yr in range(YearNumber):
        SL_wTdelay_CISM_NCA_R0_RCP85[i,yr] = SL_wTdelay_CISM_NCA_R0_RCP85[i,yr]-SL_wTdelay_CISM_NCA_R0_RCP85_offset
        SL_wTdelay_CISM_NCA_R1_RCP85[i,yr] = SL_wTdelay_CISM_NCA_R1_RCP85[i,yr]-SL_wTdelay_CISM_NCA_R1_RCP85_offset
        SL_wTdelay_CISM_NCA_R2_RCP85[i,yr] = SL_wTdelay_CISM_NCA_R2_RCP85[i,yr]-SL_wTdelay_CISM_NCA_R2_RCP85_offset
        SL_wTdelay_CISM_NCA_R3_RCP85[i,yr] = SL_wTdelay_CISM_NCA_R3_RCP85[i,yr]-SL_wTdelay_CISM_NCA_R3_RCP85_offset
        SL_wTdelay_CISM_NCA_R4_RCP85[i,yr] = SL_wTdelay_CISM_NCA_R4_RCP85[i,yr]-SL_wTdelay_CISM_NCA_R4_RCP85_offset
        SL_wTdelay_CISM_NCA_R5_RCP85[i,yr] = SL_wTdelay_CISM_NCA_R5_RCP85[i,yr]-SL_wTdelay_CISM_NCA_R5_RCP85_offset

SL_wTdelay_CISM_NCA_SU_RCP85 = SL_wTdelay_CISM_NCA_R1_RCP85+SL_wTdelay_CISM_NCA_R2_RCP85+SL_wTdelay_CISM_NCA_R3_RCP85+SL_wTdelay_CISM_NCA_R4_RCP85+SL_wTdelay_CISM_NCA_R5_RCP85

SL_wTdelay_CISM_NCA_SU_RCP85_50pc = np.percentile(SL_wTdelay_CISM_NCA_SU_RCP85, 50, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTdelay_CISM_NCA_SU_RCP85_83pc = np.percentile(SL_wTdelay_CISM_NCA_SU_RCP85, 83.33, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTdelay_CISM_NCA_SU_RCP85_17pc = np.percentile(SL_wTdelay_CISM_NCA_SU_RCP85, 16.66, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTdelay_CISM_NCA_SU_RCP85_95pc = np.percentile(SL_wTdelay_CISM_NCA_SU_RCP85, 5, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTdelay_CISM_NCA_SU_RCP85_05pc = np.percentile(SL_wTdelay_CISM_NCA_SU_RCP85, 95, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTdelay_CISM_NCA_SU_RCP85_99pc = np.percentile(SL_wTdelay_CISM_NCA_SU_RCP85, 99, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTdelay_CISM_NCA_SU_RCP85_01pc = np.percentile(SL_wTdelay_CISM_NCA_SU_RCP85, 1, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)

print(max(SL_wTdelay_CISM_NCA_SU_RCP85_99pc))
```
