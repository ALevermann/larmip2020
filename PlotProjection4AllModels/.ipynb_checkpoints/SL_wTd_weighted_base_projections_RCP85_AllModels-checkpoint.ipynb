```python
import numpy as np
import netCDF4 as nc
import matplotlib.pylab as plt
import imp
import csv
import pandas as pd
from io import StringIO
```

```python
fname = "../ObservationalData/imbie_dataset-2018_06_08.dat" # File to read
df = pd.read_csv(fname,sep='\t',index_col=-1,header=None)
ObsData = np.array(df.values)
print(ObsData[-1,:])
ObsTime = ObsData[:,0]
ObsSL = ObsData[:,3]
ObsUncertainty = ObsData[:,4]
#print(ObsData[:,0])
plt.plot(ObsTime,ObsSL,'k')
plt.plot(ObsTime,ObsSL-ObsUncertainty,':k')
plt.plot(ObsTime,ObsSL+ObsUncertainty,':k')

#ObsDSL = ObsData[:,4]
#print(ObsTime[0],ObsTime[-1],ObsTime[-1]-ObsTime[0],ObsSL[-1]-ObsSL[0])
SLObsVal9216 = (ObsSL[-1]-ObsSL[0])/1000
SLObsUnc9216 = ObsUncertainty[-1]/1000
#print(SLObsVal9216,SLObsUnc9216)


fname = "../ObservationalData/imbie_dataset-2018_06_08_WAIS.dat" # File to read
df = pd.read_csv(fname,sep='\t',header=None)
ObsDataWAIS = np.array(df.values)
print(ObsDataWAIS[-1,:])
ObsTime = ObsDataWAIS[:,0]
ObsSL_WAIS = ObsDataWAIS[:,3]
ObsUncertainty_WAIS = ObsDataWAIS[:,4]
plt.plot(ObsTime,ObsSL_WAIS,'r')
plt.plot(ObsTime,ObsSL_WAIS-ObsUncertainty_WAIS,':r')
plt.plot(ObsTime,ObsSL_WAIS+ObsUncertainty_WAIS,':r')
SLObsVal9216_WAIS = (ObsSL_WAIS[-1]-ObsSL_WAIS[0])/1000
SLObsUnc9216_WAIS = ObsUncertainty_WAIS[-1]/1000


fname = "../ObservationalData/imbie_dataset-2018_06_08_EAIS.dat" # File to read
df = pd.read_csv(fname,sep='\t',header=None)
ObsDataEAIS = np.array(df.values)
print(ObsDataEAIS[-1,:])
ObsTime = ObsDataEAIS[:,0]
ObsSL_EAIS = ObsDataEAIS[:,3]
ObsUncertainty_EAIS = ObsDataEAIS[:,4]
plt.plot(ObsTime,ObsSL_EAIS,'g')
plt.plot(ObsTime,ObsSL_EAIS-ObsUncertainty_EAIS,':g')
plt.plot(ObsTime,ObsSL_EAIS+ObsUncertainty_EAIS,':g')
SLObsVal9216_EAIS = (ObsSL_EAIS[-1]-ObsSL_EAIS[0])/1000
SLObsUnc9216_EAIS = ObsUncertainty_EAIS[-1]/1000


fname = "../ObservationalData/imbie_dataset-2018_06_08_PEN.dat" # File to read
df = pd.read_csv(fname,sep='\t',header=None)
ObsDataPEN = np.array(df.values)
print(ObsDataPEN[-1,:])
ObsTime = ObsDataPEN[:,0]
ObsSL_PEN = ObsDataPEN[:,3]
ObsUncertainty_PEN = ObsDataPEN[:,4]
plt.plot(ObsTime,ObsSL_PEN,'b')
plt.plot(ObsTime,ObsSL_PEN-ObsUncertainty_PEN,':b')
plt.plot(ObsTime,ObsSL_PEN+ObsUncertainty_PEN,':b')
SLObsVal9216_PEN = (ObsSL_PEN[-1]-ObsSL_PEN[0])/1000
SLObsUnc9216_PEN = ObsUncertainty_PEN[-1]/1000

```

```python
# Read data
# AISM_VUB

fname="../ComputeProjection4OneIceModel/EnsembleSingleModelProjections/SL_wTd_nos_base_AISM_VUB_RCP85.nc"
ncf = nc.Dataset(fname, "r")

Time = ncf.variables["Time"][:]
SL_wTd_nos_base_AISM_VUB_R0_RCP85 = ncf.variables["Antarctica"][:]
SL_wTd_nos_base_AISM_VUB_R1_RCP85 = ncf.variables["EAIS"][:]
SL_wTd_nos_base_AISM_VUB_R2_RCP85 = ncf.variables["Ross"][:]
SL_wTd_nos_base_AISM_VUB_R3_RCP85 = ncf.variables["Amundsen"][:]
SL_wTd_nos_base_AISM_VUB_R4_RCP85 = ncf.variables["Weddell"][:]
SL_wTd_nos_base_AISM_VUB_R5_RCP85 = ncf.variables["Peninsula"][:]
SL_wTd_nos_base_AISM_VUB_SU_RCP85 = SL_wTd_nos_base_AISM_VUB_R1_RCP85+SL_wTd_nos_base_AISM_VUB_R2_RCP85+SL_wTd_nos_base_AISM_VUB_R3_RCP85+SL_wTd_nos_base_AISM_VUB_R4_RCP85+SL_wTd_nos_base_AISM_VUB_R5_RCP85
ncf.close()

SL_wTd_nos_base_AISM_VUB_SU_RCP85_50pc = np.percentile(SL_wTd_nos_base_AISM_VUB_SU_RCP85, 50, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTd_nos_base_AISM_VUB_SU_RCP85_83pc = np.percentile(SL_wTd_nos_base_AISM_VUB_SU_RCP85, 83.33, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTd_nos_base_AISM_VUB_SU_RCP85_17pc = np.percentile(SL_wTd_nos_base_AISM_VUB_SU_RCP85, 16.66, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTd_nos_base_AISM_VUB_SU_RCP85_95pc = np.percentile(SL_wTd_nos_base_AISM_VUB_SU_RCP85, 5, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTd_nos_base_AISM_VUB_SU_RCP85_05pc = np.percentile(SL_wTd_nos_base_AISM_VUB_SU_RCP85, 95, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTd_nos_base_AISM_VUB_SU_RCP85_99pc = np.percentile(SL_wTd_nos_base_AISM_VUB_SU_RCP85, 99, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
SL_wTd_nos_base_AISM_VUB_SU_RCP85_01pc = np.percentile(SL_wTd_nos_base_AISM_VUB_SU_RCP85, 1, axis=0, out=None, overwrite_input=False, interpolation='linear', keepdims=False)

EnsembleSize=len(SL_wTd_nos_base_AISM_VUB_R0_RCP85[:,1])
print(EnsembleSize,Time[92],Time[117])
Weight_wTd_nos_base_AISM_VUB_R0=np.zeros(EnsembleSize)
Weight_wTd_nos_base_AISM_VUB_R1=np.zeros(EnsembleSize)
Weight_wTd_nos_base_AISM_VUB_R2=np.zeros(EnsembleSize)
Weight_wTd_nos_base_AISM_VUB_R5=np.zeros(EnsembleSize)
Weight_wTd_nos_base_AISM_VUB_R0_sum = 0
Weight_wTd_nos_base_AISM_VUB_R1_sum = 0
Weight_wTd_nos_base_AISM_VUB_R2_sum = 0
Weight_wTd_nos_base_AISM_VUB_R5_sum = 0
for i in range(len(SL_wTd_nos_base_AISM_VUB_R0_RCP85[:,1])):
    sldiff = SL_wTd_nos_base_AISM_VUB_R0_RCP85[i,117]-SL_wTd_nos_base_AISM_VUB_R0_RCP85[i,92]-SLObsVal9216
    weight = np.exp(-sldiff/SLObsUnc9216*sldiff/2)
    Weight_wTd_nos_base_AISM_VUB_R0[i] = weight
    Weight_wTd_nos_base_AISM_VUB_R0_sum = Weight_wTd_nos_base_AISM_VUB_R0_sum + weight

    sldiff = SL_wTd_nos_base_AISM_VUB_R1_RCP85[i,117]-SL_wTd_nos_base_AISM_VUB_R1_RCP85[i,92]-SLObsVal9216_EAIS
    weight = np.exp(-sldiff/SLObsUnc9216_EAIS*sldiff/2)
    Weight_wTd_nos_base_AISM_VUB_R1[i] = weight
    Weight_wTd_nos_base_AISM_VUB_R1_sum = Weight_wTd_nos_base_AISM_VUB_R1_sum + weight

    sldiff = SL_wTd_nos_base_AISM_VUB_R2_RCP85[i,117]-SL_wTd_nos_base_AISM_VUB_R2_RCP85[i,92]
    sldiff = sldiff+SL_wTd_nos_base_AISM_VUB_R3_RCP85[i,117]-SL_wTd_nos_base_AISM_VUB_R3_RCP85[i,92]
    sldiff = sldiff+SL_wTd_nos_base_AISM_VUB_R4_RCP85[i,117]-SL_wTd_nos_base_AISM_VUB_R4_RCP85[i,92]-SLObsVal9216_WAIS
    weight = np.exp(-sldiff/SLObsUnc9216_WAIS*sldiff/2)
    Weight_wTd_nos_base_AISM_VUB_R2[i] = weight
    Weight_wTd_nos_base_AISM_VUB_R2_sum = Weight_wTd_nos_base_AISM_VUB_R2_sum + weight

    sldiff = SL_wTd_nos_base_AISM_VUB_R5_RCP85[i,117]-SL_wTd_nos_base_AISM_VUB_R5_RCP85[i,92]-SLObsVal9216_PEN
    weight = np.exp(-sldiff/SLObsUnc9216_PEN*sldiff/2)
    Weight_wTd_nos_base_AISM_VUB_R5[i] = weight
    Weight_wTd_nos_base_AISM_VUB_R5_sum = Weight_wTd_nos_base_AISM_VUB_R5_sum + weight


plt.plot(Weight_wTd_nos_base_AISM_VUB_R0,'.k')
plt.plot(Weight_wTd_nos_base_AISM_VUB_R1,'.g')
plt.plot(Weight_wTd_nos_base_AISM_VUB_R2,'.r')
plt.plot(Weight_wTd_nos_base_AISM_VUB_R5,'.b')
print(Weight_wTd_nos_base_AISM_VUB_R0_sum)

#normalizing
#Weight_wTd_nos_base_AISM_VUB_R0[:] = [x / Weight_wTd_nos_base_AISM_VUB_R0_sum for x in Weight_wTd_nos_base_AISM_VUB_R0]
#Weight_wTd_nos_base_AISM_VUB_R1[:] = [x / Weight_wTd_nos_base_AISM_VUB_R1_sum for x in Weight_wTd_nos_base_AISM_VUB_R1]
#Weight_wTd_nos_base_AISM_VUB_R2[:] = [x / Weight_wTd_nos_base_AISM_VUB_R2_sum for x in Weight_wTd_nos_base_AISM_VUB_R2]
#Weight_wTd_nos_base_AISM_VUB_R5[:] = [x / Weight_wTd_nos_base_AISM_VUB_R5_sum for x in Weight_wTd_nos_base_AISM_VUB_R5]

# write weights
ncfile = nc.Dataset('ModelWeights/Weights_AISM_VUB.nc','w', format='NETCDF4')
ncfile.createDimension('Emember', None)

Weight_AISM_VUB_R0 = ncfile.createVariable('Weight for Antarctica', 'f4', ('Emember'))
Weight_AISM_VUB_R1 = ncfile.createVariable('Weight for EAIS', 'f4', ('Emember'))
Weight_AISM_VUB_R2 = ncfile.createVariable('Weight for WAIS', 'f4', ('Emember'))
Weight_AISM_VUB_R5 = ncfile.createVariable('Weight for Peninsula', 'f4', ('Emember'))

Weight_AISM_VUB_R0[:] = Weight_wTd_nos_base_AISM_VUB_R0
Weight_AISM_VUB_R1[:] = Weight_wTd_nos_base_AISM_VUB_R1
Weight_AISM_VUB_R2[:] = Weight_wTd_nos_base_AISM_VUB_R2
Weight_AISM_VUB_R5[:] = Weight_wTd_nos_base_AISM_VUB_R5

Weight_AISM_VUB_R0.units = 'one'
Weight_AISM_VUB_R1.units = 'one'
Weight_AISM_VUB_R2.units = 'one'
Weight_AISM_VUB_R5.units = 'one'


ncfile.close()
```
